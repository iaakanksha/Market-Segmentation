```python
data1 = df1.copy()
```

```python
data1.head()
```

```python
from sklearn.preprocessing import LabelEncoder

labelencoder_X = LabelEncoder()
for i in data1.columns:
    data1[i] = labelencoder_X.fit_transform(data1[i])
```

```python
data1.head()
```

```python
data1.shape
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset

X = data1.iloc[:, [0,1]].values

# Using the elbow method to find the optimal number of clusters
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 23)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 20)
y_kmeans = kmeans.fit_predict(data1[['locations','roles','experience']])
kmeans.fit(data1)
df1['Cluster'] = kmeans.labels_
```

```python
# Visualising the clusters
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'cyan', label = 'Cluster 3')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of market analysis')
plt.xlabel("Experience")
plt.ylabel("Roles")
plt.legend()
plt.show()
```

```python
import plotly.graph_objects as go
PLOT = go.Figure()
for C in list(df1.Cluster.unique()):
    
    PLOT.add_trace(go.Scatter3d(x = df1[df1.Cluster == C]['locations'],
                                y = df1[df1.Cluster == C]['experience'],
                                z = df1[df1.Cluster == C]['roles'],
                                mode = 'markers', marker_size = 8, marker_line_width = 1,
                                name = 'Cluster ' + str(C)))


PLOT.update_layout(width = 800, height = 800, autosize = True, showlegend = True,
                   scene = dict(xaxis=dict(title = 'Location', titlefont_color = 'pink'),
                                yaxis=dict(title ='Experience' , titlefont_color = 'pink'),
                                zaxis=dict(title = 'Role', titlefont_color = 'pink')),
                   font = dict(family = "Gilroy", color  = 'black', size = 10))
```

# Model Creation

```python
data2 = df.copy()
data2.head()
```

```python
from sklearn.preprocessing import LabelEncoder

labelencoder_X = LabelEncoder()
for i in data2.columns:
    data2[i] = labelencoder_X.fit_transform(data2[i])
```

```python
data2.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset

X = data2.iloc[:, [0,1]].values

# Using the elbow method to find the optimal number of clusters
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Training the K-Means model on the dataset
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 10)
y_kmeans = kmeans.fit_predict(data2[['location','job_desig','experience']])
kmeans.fit(data2)
df['Cluster'] = kmeans.labels_
```

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'cyan', label = 'Cluster 3')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('Location')
plt.ylabel('Job designation')
plt.legend()
plt.show()
```

```python
import plotly.graph_objects as go
PLOT = go.Figure()
for C in list(df.Cluster.unique()):
    
    PLOT.add_trace(go.Scatter3d(x = df[df.Cluster == C]['location'],
                                y = df[df.Cluster == C]['job_desig'],
                                z = df[df.Cluster == C]['experience'],
                                mode = 'markers', marker_size = 8, marker_line_width = 1,
                                name = 'Cluster ' + str(C)))


PLOT.update_layout(width = 800, height = 800, autosize = True, showlegend = True,
                   scene = dict(xaxis=dict(title = 'location', titlefont_color = 'pink'),
                                yaxis=dict(title ='job_desig' , titlefont_color = 'pink'),
                                zaxis=dict(title = 'experience', titlefont_color = 'pink')),
                   font = dict(family = "Gilroy", color  = 'black', size = 10))
```
